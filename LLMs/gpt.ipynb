```python
import torch
```

```python
with open('./Shakespeare.txt', 'r') as file:
    data = file.read()

vocab = sorted(set(data))

#Encoding and decoding
enc = {ch:i for i,ch in enumerate(vocab)}
dec = {i:ch for i,ch in enumerate(vocab)}
encode = lambda s : [enc[i] for i in s]
decode = lambda t : ''.join([dec[c] for c in t])

encode(data)
```

```python
#Data splitting
encoded_data = encode(data)

final_data = torch.tensor(encoded_data,dtype=torch.long)
train_data = final_data[:round(len(final_data)*0.8)]
test_data = final_data[round(len(final_data)*0.8):]

block_size = 8
batch_size = 4
```
